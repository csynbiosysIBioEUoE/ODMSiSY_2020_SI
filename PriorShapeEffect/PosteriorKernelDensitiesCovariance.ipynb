## Compute matrix norms to assess difference

```python
from numpy import linalg as LA
```

Correlation matrices

```python
# Norm between correlation matrix difference
LA.norm(POS1.corr()-POS2.corr())
```

```python
# Norm of correlation matrix for original prior
LA.norm(POS1.corr())
```

```python
# Norm of correlation matrix for weakly informative prior
LA.norm(POS2.corr())
```

Covariance matrices

```python
# Norm between covariance matrix difference
LA.norm(POS1.cov()-POS2.cov())
```

```python
# Norm of covariance matrix for original prior
LA.norm(POS1.cov())
```

```python
# Norm of covariance matrix for weakly informative prior
LA.norm(POS2.cov())
```
